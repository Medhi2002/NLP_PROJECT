```python
import pandas as pd

def load_dataset(base_path, split):
    """
    Loads text and labels from a specified split (Train or Test).

    Args:
    - base_path (str): Path to the dataset.
    - split (str): 'Train' or 'Test'.

    Returns:
    - pd.DataFrame: A DataFrame with 'text' and 'label' columns.
    """
    data = []
    split_path = os.path.join(base_path, split)
    categories = os.listdir(split_path)

    for label, category in enumerate(categories):
        category_path = os.path.join(split_path, category)
        if not os.path.isdir(category_path):
            continue  # Skip non-folder items
        files = os.listdir(category_path)
        for file_name in files:
            file_path = os.path.join(category_path, file_name)
            if os.path.isdir(file_path):  # Skip subdirectories
                continue
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    text = f.read()
                    data.append({'text': text, 'label': label})
            except Exception as e:
                print(f"Error reading file {file_path}: {e}")

    return pd.DataFrame(data)

# Define dataset paths
base_path = "/content/arabiya_dataset/arabiya"  # Adjust as necessary

# Load Train and Test datasets
train_data = load_dataset(base_path, "Train")
test_data = load_dataset(base_path, "Test")

print("Train Data Sample:")
print(train_data.head())

print("Test Data Sample:")
print(test_data.head())
```

```python
def remove_metadata(text):
    # Remove lines starting with specific metadata fields
    lines = text.split('\n')
    filtered_lines = [line for line in lines if not line.startswith(('Title', 'Source'))]
    return ' '.join(filtered_lines).strip()

train_data['text'] = train_data['text'].apply(remove_metadata)
test_data['text'] = test_data['text'].apply(remove_metadata)
```

```python
def normalize_numbers(text):
    arabic_numbers = str.maketrans('0123456789', '٠١٢٣٤٥٦٧٨٩')
    return text.translate(arabic_numbers)

train_data['text'] = train_data['text'].apply(normalize_numbers)
test_data['text'] = test_data['text'].apply(normalize_numbers)
```

```python
import re
def remove_whitespace(text):
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces/newlines with single space
    return text.strip()

train_data['text'] = train_data['text'].apply(remove_whitespace)
test_data['text'] = test_data['text'].apply(remove_whitespace)
```

```python
import re

def remove_english_words(text):
    """
    Removes English words, abbreviations, and alphanumeric tokens from the text.
    Args:
    - text (str): The raw input text.
    Returns:
    - str: The cleaned text with English words removed.
    """
    # Remove English words, numbers, and alphanumeric tokens
    text = re.sub(r'[a-zA-Z0-9]+', '', text)
    return text
train_data['text'] = train_data['text'].apply(remove_english_words)
test_data['text'] = test_data['text'].apply(remove_english_words)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

def remove_stopwords(text):
    arabic_stopwords = set(stopwords.words("arabic"))
    words = text.split()
    filtered_words = [word for word in words if word not in arabic_stopwords]
    return ' '.join(filtered_words)

train_data['text'] = train_data['text'].apply(remove_stopwords)
test_data['text'] = test_data['text'].apply(remove_stopwords)

```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class DifferentialAttention(nn.Module):
    def __init__(self, d_model, num_heads, lambda_init=0.8):
        super(DifferentialAttention, self).__init__()
        self.num_heads = num_heads
        self.d_head = d_model // num_heads
        self.lambda_init = lambda_init

        # Linear layers for Q, K, V projections
        self.query = nn.Linear(d_model, 2 * d_model)
        self.key = nn.Linear(d_model, 2 * d_model)
        self.value = nn.Linear(d_model, d_model)

        # Learnable scalars
        self.lambda_q1 = nn.Parameter(torch.randn(self.d_head))
        self.lambda_k1 = nn.Parameter(torch.randn(self.d_head))
        self.lambda_q2 = nn.Parameter(torch.randn(self.d_head))
        self.lambda_k2 = nn.Parameter(torch.randn(self.d_head))

    def forward(self, X):
        # Projections
        Q = self.query(X)
        K = self.key(X)
        V = self.value(X)

        # Split into Q1, Q2, K1, K2
        Q1, Q2 = Q.chunk(2, dim=-1)
        K1, K2 = K.chunk(2, dim=-1)

        # Attention scores
        scores1 = torch.matmul(Q1, K1.transpose(-2, -1)) / (self.d_head ** 0.5)
        scores2 = torch.matmul(Q2, K2.transpose(-2, -1)) / (self.d_head ** 0.5)

        # Compute lambda
        lambda_val = (
            torch.exp(torch.dot(self.lambda_q1, self.lambda_k1)) -
            torch.exp(torch.dot(self.lambda_q2, self.lambda_k2)) +
            self.lambda_init
        )

        # Differential attention
        diff_attn = F.softmax(scores1, dim=-1) - lambda_val * F.softmax(scores2, dim=-1)
        output = torch.matmul(diff_attn, V)

        return output
```

```python
from transformers import BertModel, BertConfig

class ModifiedBertModel(nn.Module):
    def __init__(self, pretrained_model_name, num_labels):
        super(ModifiedBertModel, self).__init__()
        self.bert = BertModel.from_pretrained(pretrained_model_name)
        self.num_labels = num_labels
        self.differential_attention = DifferentialAttention(
            d_model=self.bert.config.hidden_size,
            num_heads=self.bert.config.num_attention_heads
        )
        self.classifier = nn.Linear(self.bert.config.hidden_size, num_labels)

    def forward(self, input_ids, attention_mask, labels=None):
        # Get BERT outputs
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        sequence_output = outputs.last_hidden_state

        # Apply differential attention
        diff_output = self.differential_attention(sequence_output)

        # Classification layer
        logits = self.classifier(diff_output[:, 0, :])  # Use [CLS] token representation
        loss = None
        if labels is not None:
            loss_fn = nn.CrossEntropyLoss()
            loss = loss_fn(logits, labels)

        return {'loss': loss, 'logits': logits}  # Ensure loss and logits are part of the dictionary
```

```python
def predict_batch_inputs(model, inputs, tokenizer, category_names):
    """
    Predict categories for a batch of input texts.

    Args:
    - model: The trained model (e.g., modified model).
    - inputs: A list of strings (input texts).
    - tokenizer: The tokenizer corresponding to the model.
    - category_names: List of category names for decoding predictions.

    Returns:
    - results: A list of dictionaries containing text, predicted category, and confidence score.
    """
    device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
    model.to(device)
    model.eval()

    # Tokenize inputs
    tokens = tokenizer(inputs, padding=True, truncation=True, max_length=512, return_tensors="pt")
    tokens.pop('token_type_ids', None)  # Remove token_type_ids if it exists
    tokens = {key: val.to(device) for key, val in tokens.items()}

    # Predict
    with torch.no_grad():
        outputs = model(**tokens)
        logits = outputs['logits']
        probabilities = F.softmax(logits, dim=-1).cpu().numpy()
        predictions = torch.argmax(logits, dim=-1).cpu().numpy()

    # Collect results
    results = []
    for i, text in enumerate(inputs):
        results.append({
            "text": text,
            "predicted_category": category_names[predictions[i]],
            "confidence_score": probabilities[i][predictions[i]]
        })

    return results
```
